```python
# Edge Detection Comparison in Image Processing

This Jupyter notebook explores and compares popular edge detection algorithms - Canny, Prewitt, Roberts, and Sobel. Using the well-known "Cameraman" image, we'll delve into the fundamental techniques employed by these algorithms to highlight edges and boundaries. The objective is to gain insights into the strengths and characteristics of each method. The code implementation utilizes Python with OpenCV and Matplotlib for a hands-on exploration. Let's uncover the world of edge detection!
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the Cameraman image
image = cv2.imread('.\Data\cameraman.jpg', cv2.IMREAD_GRAYSCALE)

# Apply different edge detection algorithms
canny_edges = cv2.Canny(image, 50, 150)
prewitt_kernelx = np.array([[-1, 0, 1], [-1, 0, 1], [-1, 0, 1]])
prewitt_kernely = np.array([[1, 1, 1], [0, 0, 0], [-1, -1, -1]])
prewitt_edges = np.abs(cv2.filter2D(image, -1, prewitt_kernelx) + cv2.filter2D(image, -1, prewitt_kernely))

roberts_kernelx = np.array([[1, 0], [0, -1]])
roberts_kernely = np.array([[0, 1], [-1, 0]])
roberts_edges = np.abs(cv2.filter2D(image, -1, roberts_kernelx) + cv2.filter2D(image, -1, roberts_kernely))

sobel_edges_x = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=3)
sobel_edges_y = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=3)
sobel_edges = np.abs(sobel_edges_x) + np.abs(sobel_edges_y)

# Plot the original and edge-detected images
plt.figure(figsize=(12, 8))

plt.subplot(231), plt.imshow(image, cmap='gray'), plt.title('Original Image')
plt.subplot(232), plt.imshow(canny_edges, cmap='gray'), plt.title('Canny Edge Detection')
plt.subplot(233), plt.imshow(prewitt_edges, cmap='gray'), plt.title('Prewitt Edge Detection')
plt.subplot(234), plt.imshow(roberts_edges, cmap='gray'), plt.title('Roberts Edge Detection')
plt.subplot(235), plt.imshow(sobel_edges, cmap='gray'), plt.title('Sobel Edge Detection')

plt.show()
```
